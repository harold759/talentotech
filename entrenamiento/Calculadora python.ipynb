# Calculadora basica

```python
print('Calculadora')
print()
num_one=float(input('Ingrese el primer numero '))
num_two=float(input('Ingrese el segundo numero '))
print()
print('MENU')
print('* Digite numero 1 para sumar')
print('* Digite numero 2 para restar')
print('* Digite numero 3 para multiplicar')
print('* Digite numero 4 para dividir')
print()
opcion=int(input('Ingrese el numero de la operacion: '))
print()
if opcion>=1 and opcion<=4:
  if opcion==1:
    result=(num_one+num_two)
    print('* El resultado de la suma es: ',result)

  if opcion==2:
    result=(num_one-num_two)
    print('* El resultado de la resta es: ',result)

  if opcion==3:
    result=(num_one*num_two)
    print('* El resultado de la multiplicación es: ',result)

  if opcion==4:
    if num_two!= 0:
      result=(num_one/num_two)
      print('* El resultado de la multiplicación es: ',result)
    else:
      print('No se puede dividir por 0 ')

else:
  print('opcion no valida digite un numero entre 1 y 4')
```

# Calculadora Mejorada

```python
import math

print('Calculadora Cientifica')
print()
opcion=0
while opcion != 9:
  print()
  print('MENU')
  print('* Digite numero 1 para sumar')
  print('* Digite numero 2 para restar')
  print('* Digite numero 3 para multiplicar')
  print('* Digite numero 4 para dividir')
  print('* Digite numero 5 para elevar')
  print('* Digite numero 6 para raiz cuadrada')
  print('* Digite numero 7 para raiz cubica')
  print('* Digite numero 8 para tangente')
  print('* Digite numero 9 para salir')
  print()
  opcion=int(input('Ingrese el numero de la operacion: '))
  print()
  if opcion>=1 and opcion<=8:
    if opcion>= 1 and opcion<=5:
      num_one=float(input('Ingrese el primer numero '))
      num_two=float(input('Ingrese el segundo numero '))

      match opcion:
        case 1:
          result=(num_one+num_two)
          print()
          print('* El resultado de la suma es: ',result)

        case 2:
          result=(num_one-num_two)
          print()
          print('* El resultado de la resta es: ',result)

        case 3:
          result=(num_one*num_two)
          print()
          print('* El resultado de la multiplicación es: ',result)

        case 4:
          if num_two!= 0:
            result=(num_one/num_two)
            print()
            print('* El resultado de la division  es: ',result)
          else:
            print()
            print('No se puede dividir por 0 ')

        case 5:
          result=pow(num_one,num_two)
          print()
          print('* El resultado de la operacion es: ',result)

    if opcion>5 and opcion<10:
      num_one=float(input('Ingrese un numero'))

      match opcion:
        case 6:
          result=math.sqrt(num_one)
          print()
          print('* El resultado de la raiz cuadrada es: ',result)

        case 7:
          result=math.cbrt(num_one)
          print()
          print('* El resultado de la raiz cubica es: ',result)

        case 8:
          result=math.tan(num_one)
          print()
          print('* El resultado de la tangente es: ',result)
  else:
    print('Saliste de la calculadora')
```
